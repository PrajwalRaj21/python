```python
df = pd.read_csv("CD.csv")
```

```python
df.head()
```

```python
car_company_counts = df['Company'].value_counts()
print(car_company_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=car_company_counts.index, y=car_company_counts.values)
plt.title('Number of Cars by Company')
plt.xlabel('Company')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

The above bar graph displays the number of cars that are available from different companies

```python
fuel_type_counts = df['FuelType'].value_counts()
print(fuel_type_counts)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(fuel_type_counts.index, fuel_type_counts.values, color='blue')
plt.title('Number of vehicles of different fuel types')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python



bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
labels = ['0-50', '51-100', '101-150', '151-200', '201-250', '251-300', '301-350', '351-400', '401-450', '451-500']

rental_price_groups = df.groupby(pd.cut(df['Rental_Price'], bins=bins, labels=labels)).size()


plt.figure(figsize=(10, 6))
rental_price_groups.plot(kind='bar', color='skyblue')
plt.title('Number of vehicles by Rental Price Range')
plt.xlabel('Rental Price Range ')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python


bins = [0,1.5,2.5,3.5,4.5,5]
labels = ['1-1.5', '1.6-2.5', '2.6-3.5', '3.6-4.5', '4.6-5']
rating_groups = df.groupby(pd.cut(df['Rating'], bins=bins, labels=labels)).size()


plt.figure(figsize=(8, 6))
rating_groups.plot(kind='bar', color='skyblue')
plt.title('Number of vehicles by Rating')
plt.xlabel('Rating')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=0)
plt.show()
```

```python
vehicle_type_counts = df['VehicleType'].value_counts()

plt.figure(figsize=(10, 6))
vehicle_type_counts.plot(kind='bar', color='skyblue')
plt.title('Number of vehicles by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

The above graph shows the different vehicles based on their types available.
